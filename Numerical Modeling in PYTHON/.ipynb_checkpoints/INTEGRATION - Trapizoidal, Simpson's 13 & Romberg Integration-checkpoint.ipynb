```python
########################
#Function
f = lambda x: math.log(x)

#Integration Limits
a = 2; b= 4
########################

ea = 100; n = 1; old = 100

print ('n \t\t ea  \t\t et \t\t integral')

while ea > 0.001:
    #Calculate Number of Segments
    h = (b-a)/n
    
    #Calculate summation term
    Sum = 0
    for i in range(1,n):
        Sum += 2 * f(a + i*h) 
        
    #Calculate Integral estimate
    integral = (h/2)*(f(a)+Sum+f(b))
    
    
    #Calculate approximate & true error
    ea = abs((integral - old)/ integral)*100
    et = abs((2.15888 - integral)/ 2.15888)*100
    
    print ('%f \t %f \t %f \t %f'% (n, ea, et, integral))
    
    #Store current estimate for next iteration
    old = integral
    
    #Increase number of trapezoids by one
    n += 1
    
    
    
```

```python
########################
#Function#
f = lambda x: math.log(x)

#Integration Limits
a = 2; b = 4
########################

ea = 100; n = 2; old = 100; ii = 1

print ('n \t\t ea  \t\t et \t\t integral')

while ea > 0.01:
    
    #Calculate Number of Segments
    h = (b-a)/n
     
    #Calculate 1st Summation term
    Sum1= 0
    for i in range(1, n, 2):
        Sum1 += 4*f(a+i*h)
    
    #Calculate 2nd Summation term
    Sum2= 0
    for i in range(2, n-1, 2):
        Sum2 += 2*f(a+i*h)          
    
    #Calcualte integral estimation    
    integral = (h/3)*(f(a)+Sum1+Sum2+f(b))
    
    #Calculate approximate & true error
    ea = abs((integral - old)/ integral)*100
    et = abs((2.15888 - integral)/ 2.15888)*100
    
    print ('%f \t %f \t %f \t %f'% (ii, ea, et, integral))
    
    #Store root for next iteration
    old = integral
    
    #Increase number of applications of simpson's by one
    n += 2
    
    ii += 1
    
```

```python
import math
import numpy as np
```

```python
########################
#Function#
f = lambda x: math.exp(x)

#Integration Limits
a = 0; b = 1
#######################################
#Trapezoidal Function
def Trap(n, a, b):
    #Calculate Number of Segments
    h = (b - a) / n
    #Calculate summation term
    Sum = 0
    for i in range(1, n):
        Sum = Sum + 2 * f(a + i * h)
    #Calculate Trapezoidal Estimate
    Trap = (h / 2) * (f(a) + Sum + f(b))
    return Trap
#######################################

#initialize
I= np.zeros((10,10))
ea = 100; n = 1; i = 1
I[1, 1] = Trap(n, a, b)

print ('Application \t ea \t\t et \t\t Integral') 

while ea > 0.001:
    #Increase the number of trapezoids
    n = 2 ** i
    
    #Next Integral Approx.
    I[i+1 , 1] = Trap(n, a, b)
    
    #romberg integration
    for k in range(2, i + 2):
        j = 2 + i - k
        I[j, k] = (4 ** (k - 1) * I[j+1, k - 1] - I[j, k - 1]) / (4 ** (k - 1) - 1)
    
    #Calculate approximate & true error
    ea = abs((I[1, i+1] - I[2, i]) / I[1, i+1]) * 100
    et = abs((1.71828 - I[1, i+1]) / 1.7182) * 100
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, I[1,i+1]))
    
    i += 1
```
