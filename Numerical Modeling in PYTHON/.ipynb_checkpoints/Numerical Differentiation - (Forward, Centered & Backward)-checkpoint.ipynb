# Finite Differences Approximation

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Finite Differance Approximation

```python
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
dy = lambda x: 0.5*x**4-0.6*x**2+0.1
dy2 = lambda x: 2*x**3-1.2*x
h = 0.05
x = 0.1

# Forward differences approximation
dff1 = (f(x+h) - f(x))/h
dff2 = (f(x+2*h) - 2*f(x+h) + f(x))/h**2
er_dff1 = ((dy(x)-dff1)/dy(x))*100
er_dff2 = ((dy2(x)-dff2)/dy2(x))*100
print('Solution by forward differences:')
print('f\'(%f) = %f'%(x,dff1))
print('Error in dff1 = {} {}'.format(round(er_dff1,3),'%'))
print('f\'\'(%f) = %f'%(x,dff2))
print('Error in dff2 = {} {}'.format(round(er_dff2,3),'%'))

# Central differences approximation
dfc1 = (f(x+h)-f(x-h))/(2*h)
dfc2 = (f(x+h)-2*f(x)+f(x-h))/h**2
er_dfc1 = ((dy(x)-dfc1)/dy(x))*100
er_dfc2 = ((dy2(x)-dfc2)/dy2(x))*100
print('\nSolution by central differences:')
print('f\'(%f) = %f'%(x,dfc1))
print('Error in dfc1 = {} {}'.format(round(er_dfc1,3),'%'))
print('f\'\'(%f) = %f'%(x,dfc2))
print('Error in dfc2 = {} {}'.format(round(er_dfc2,3),'%'))

# High Accuracy Central differences approximation
Hfc1 = (-f(x + 2 * h) + 8 * f(x + h) - 8 * f(x - h) + f(x - 2 * h)) / (12 * h)
Hfc2 = (-f(x + 2 * h) + 16 * f(x + h) - 30 * f(x) + 16 * f(x - h) - f(x - 2 * h)) / (12 * h ** 2)
er_Hfc1 = ((dy(x)-Hfc1)/dy(x))*100
er_Hfc2 = ((dy2(x)-Hfc2)/dy2(x))*100
print('\nSolution by Hight Accuracy Central:')
print('f\'(%f) = %f'%(x,Hfc1))
print('Error in Hfc1 = {} {}'.format(round(er_Hfc1,3),'%'))
print('f\'\'(%f) = %f'%(x,Hfc2))
print('Error in Hfc1 = {} {}'.format(round(er_Hfc2,3),'%'))


# Backward differences approximation
dfb1 = (f(x)-f(x-h))/h
dfb2 = (f(x)-2*f(x-h)+f(x-2*h))/h**2
er_dfb1 = ((dy(x)-dfb1)/dy(x))*100
er_dfb2 = ((dy2(x)-dfb2)/dy2(x))*100
print('\nSolution by backward differences:')
print('f\'(%f) = %f'%(x,dfb1))
print('Error in dfb1 = {} {}'.format(round(er_dfb1,3),'%'))
print('f\'\'(%f) = %f'%(x,dfb2))
print('Error in dfb1 = {} {}'.format(round(er_dfb2,3),'%'))

```

## Forward FDA (Plot)

```python
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
dff1 = (f(x+h) - f(x))/h
dff2 = (f(x+2*h) - 2*f(x+h) + f(x))/h**2

h = 0.5

# Plotting results
x = np.linspace(0,1,11)
plt.plot(x,f(x),'-k', x,dfc1,'--b', x,dfc2,'-.r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','f \'(x)','f \'\'(x)'])
plt.grid()
plt.show()
```

## Backward FDA (Plot)

```python
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
dfb1 = (f(x)-f(x-h))/h
dfb2 = (f(x)-2*f(x-h)+f(x-2*h))/h**2

h = 0.5

# Plotting results
x = np.linspace(0,1,11)
plt.plot(x,f(x),'-k', x,dfc1,'--b', x,dfc2,'-.r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','f \'(x)','f \'\'(x)'])
plt.grid()
plt.show()
```
