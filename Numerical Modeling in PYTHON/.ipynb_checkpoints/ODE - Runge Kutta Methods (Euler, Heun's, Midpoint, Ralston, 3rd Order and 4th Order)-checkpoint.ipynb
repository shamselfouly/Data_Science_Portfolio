## Problem Information 

### Function(s)

```python
#Functions
dy = lambda x,y: y+3
f = lambda x: math.exp(x-2) - 3
```

### Integration Limits & Number of Calculations

```python
#Set integraxion limits & Calculate number of Calculations
xi = 2; xf = 4
h = 0.2
n = int((xf - xi) / h)
```

## Euler

```python
#Initial Condition
x = 2; y=-2

print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot = []; y_euler = []; y_analytical = []

#Euler's Method
for i in range(1, n+1):
    x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
    y = y + dy(x, y) * h
    x = x + h
    print ('%f \t %f \t %f'% (x, y, f(x))) 
    
x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_euler,'ro', x_plot, y_analytical)
plt.legend(['Euler','Analytical'])
plt.xlabel('x'); plt.ylabel('y')
```

## Heun's

```python
#Initialization
x = 2; y=-2

print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot = []; y_heuns = []; y_analytical = []

#Heun's Method
for i in range(1, n+1):
    x_plot.append(x); y_heuns.append(y); y_analytical.append(f(x))
    #Apply Euler Method to Get y at end of interval
    dy1 = dy(x,y) 
    ye = y + dy1 * h
    #Calculaxe derivative ax end point
    dy2 = dy(x+h, ye)
    #Calculate new y estimation
    y = y + (dy1 + dy2)/2 * h
    x = x + h
    print ('%f \t %f \t %f'% (x, y, f(x))) 
    
x_plot.append(x); y_heuns.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_heuns,'ro', x_plot, y_analytical)
plt.legend(["Heun's",'Analytical'])
plt.xlabel('x'); plt.ylabel('y')
```

## Midpoint

```python
#Initialization
x = 2; y=-2

print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot = []; y_midpoint = []; y_analytical = []

#Midpoint Method
for i in range(1, n+1):
    x_plot.append(x); y_midpoint.append(y); y_analytical.append(f(x))
    #Apply Euler Method xo Get y at end of interval
    dy1 = dy(x,y) 
    ym = y + dy1 * h/2
    #Calculaxe derivative ax end point
    dym = dy(x+h/2, ym)
    #Calculate new y estimation
    y = y + dym * h
    x = x + h    
    print ('%f \t %f \t %f'% (x, y, f(x))) 
    
x_plot.append(x); y_midpoint.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_midpoint,'ro', x_plot, y_analytical)
plt.legend(['Midpoint','Analytical'])
plt.xlabel('x'); plt.ylabel('y')
```

## Ralston

```python
#Initialization
x = 2; y=-2

print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot = []; y_ralston = []; y_analytical = []

#Ralston's Method
for i in range(1, n+1):
    x_plot.append(x); y_ralston.append(y); y_analytical.append(f(x))
    #Apply Euler Method xo Get y at end of interval
    dy1 = dy(x,y) 
    y34 = y + dy1 * 3 * h / 4
    #Calculaxe derivative ax end point
    dy34 = dy(x+ 3* h/ 4, y34)
    #Calculate new y estimation
    y = y + (1 / 3 * dy1 + 2 / 3 * dy34) * h
    #Increase x by step size to calculate next y estimate
    x = x + h    
    print ('%f \t %f \t %f'% (x, y, f(x))) 
    
x_plot.append(x); y_ralston.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_ralston,'ro', x_plot, y_analytical)
plt.legend(['Ralston','Analytical'])
plt.xlabel('x'); plt.ylabel('y')
```

## 3rd order Runge Kutta 

```python
#Initialization
x = 2; y=-2

#Visualization Code initialization
print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot= []; y_RK3 = []; y_analytical = []

#RK3 Method
for i in range(1, n+1):   
    x_plot.append(x); y_RK3.append(y); y_analytical.append(f(x))
    #Calculate Derevative approximations
    k1 = dy(x,y)  
    k2 = dy(x+h/2, y + k1 * h / 2)
    k3 = dy(x+h/2,y + (- k1 + 2*k2)*h)        
    #Calculate new y estimation
    y = y + 1 / 6 * (k1 + 4*k2 + k3) * h      
    #Increase x by step size to calculate next y estimate
    x = x + h
    print ('%f \t %f \t %f'% (x, y, f(x))) 
    
#Visulaization Code
x_plot.append(x); y_RK3.append(y); y_analytical.append(f(x))
plt.plot(x_plot,y_RK3,'ro',x_plot,y_analytical)
plt.xlabel('x'); plt.ylabel('y')
plt.legend(["RK3", 'Analytical'])
```

## 4th order Runge Kutta 

```python
#Initialization
x = 2; y=-2

#Visualization Code initialization
print ('x  \t\t y \t\t f(x)'); print ('%f \t %f \t %f'% (x, y, f(x))) 
x_plot= []; y_RK4 = []; y_analytical = []

#RK4 Method
for i in range(1, n+1):   
    x_plot.append(x); y_RK4.append(y); y_analytical.append(f(x))    
    #Euler's Method
    k1 = dy(x,y) 
    k2 = dy(x+h/2, y + k1 * h/2)    
    k3 = dy(x+h/2, y + k2*h/2)
    k4 = dy(x+h, y + k3*h)    
    #Calculate new y estimation
    y = y + 1 / 6 * (k1 + 2 * k2 + 2 * k3 + k4) * h   
    #Increase x by step size to calculate next y estimate
    x = x + h
    print ('%f \t %f \t %f'% (x, y, f(x)))
    
#Visulaization Code
x_plot.append(x); y_RK4.append(y); y_analytical.append(f(x))
plt.plot(x_plot,y_RK4,'ro',x_plot,y_analytical)
plt.xlabel('x'); plt.ylabel('y')
plt.legend(["RK4", 'Analytical'])
```
