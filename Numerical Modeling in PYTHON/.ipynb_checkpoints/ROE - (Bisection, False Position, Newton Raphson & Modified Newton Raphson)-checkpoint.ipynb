## Bisection

```python
########################
#Function#
f = lambda x: x**2 - x -2 

#Initial Guesses (Upper & Lower limits)
xl = -1.5; xu = 0
########################

#Initialize Variables
ea = 100; i = 1 ; xold = xu
print ('i  \t\t ea \t\t et \t\t xr') 

while ea > 0.001:
    #Bisection Root Finding Formula
    xr = (xu + xl)/2
    
    #Identify the section where the root exists
    if f(xr) * f(xl) < 0:
        xu = xr
    else:
        xl = xr 
        
    #Calculate approximate & True Error
    ea = abs((xr - xold) / xr) * 100
    et = abs((-1 - xr) / -1) * 100
    
    #Store root for next iteration
    xold = xr
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, xr))
    i += 1  
```

## False Position

```python
########################
#Function#
f = lambda x: x**2 - x -2 

#Initial Guesses (Upper & Lower limits)
xl = -1.5; xu = 0
########################

#Initialize Variables
ea = 100; i = 1 ; xold = xu
print ('i  \t\t ea \t\t et \t\t xr') 

while ea > 0.001:
    #False Position Root Finding Formula
    xr = xu - (f(xu) * (xl - xu) / (f(xl) - f(xu)))
    
    #Identify the section where the root exists
    if f(xr) * f(xl) < 0:
        xu = xr
    else:
        xl = xr 
        
    #Calculate approximate & True Error
    ea = abs((xr - xold) / xr) * 100
    et = abs((-1 - xr) / -1) * 100
    
    #Store root for next iteration
    xold = xr
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, xr))
    i += 1  
```

```python
#########################
#Function
f = lambda x: x**2 - x -2 
dy = lambda x: 2*x -1

#Initial guess
xi = 1
#########################

ea = 100; i = 1
print ('i  \t\t ea \t\t et \t\t xr') 

while ea > 0.001:
    #Newton Raphson root finding formula
    xr = xi - (f(xi) / dy(xi))
    
    #Calculate approximate error
    ea = abs((xr - xi) / xr) * 100
    et = abs((2 - xr) / 2) * 100
    
    #Store root for next iteration
    xi = xr
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, xr))
    i += 1
```

```python
#########################
#Function
f = lambda x: x**2 - x -2 

#Initial guess & Step Size
xi = 1; h = 0.5
#########################

ea = 100; i = 1
print ('i  \t\t ea \t\t et \t\t xr')

while ea > 0.001:
    #Secant root finding formula
    xr = xi - (f(xi) * ((xi - h) - xi)) / (f(xi - h) - f(xi))
    
    #Calculate approximate error
    ea = abs((xr - xi) / xr) * 100
    et = abs((2 - xr) / 2) * 100
    
    #Store root for next iteration
    xi = xr
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, xr))
    i += 1
```

### Modified Newton Raphson

```python
#########################
#Functions#
f = lambda x: x**3 - 7*x**2 +15*x - 9 
dy = lambda x: 3*x**2 -14*x+15
dy2 = lambda x: 6*x-14

#Initial guess
xi = 2.5
#########################

ea = 100; i = 1
print ('i  \t\t ea \t\t et \t\t xr')

while ea > 0.1:
    #Modified Newton Raphson root finding formula
    xr = xi - (f(xi) * dy(xi)) / (dy(xi) ** 2 - (f(xi) * dy2(xi)))
    
    #Calculate approximate error
    ea = abs((xr - xi) / xr) * 100
    et = abs((3 - xr) / 3) * 100
    
    #Store root for next iteration
    xi = xr
    
    print ('%f \t %f \t %f \t %f'% (i, ea, et, xr))
    i = i + 1
```
