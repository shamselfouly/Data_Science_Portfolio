```python
flights_sml = flights.filter(regex = "^year$|^month$|^day$|delay$|^distance$|^air_time$")
```

#### Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers. Convert them to a more convenient representation of number of minutes since midnight.

```python
flights.filter(['dep_time', 'sched_dep_time']).assign(min_midnight_dep = lambda x : x.dep_time // 100 * 60 + x.dep_time % 100, min_midnight_sch = lambda x : x.sched_dep_time // 100 * 60 + x.sched_dep_time % 100)
```

#### Compare air_time with arr_time - dep_time. What do you expect to see? What do you see? What do you need to do to fix it?

```python
flights.filter(['arr_time', 'dep_time', 'air_time']).assign(arr_time_minus_dep_time = lambda x : x.arr_time - x.dep_time, diff = lambda x : x.arr_time_minus_dep_time - x.air_time)
```

#### Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?

```python
flights.filter(['dep_time', 'sched_dep_time', 'dep_delay']).assign(dep_time_min = lambda x : x.dep_time // 100 * 60 + x.dep_time % 100, sched_dep_time_min = lambda x : x.sched_dep_time // 100 * 60 + x.sched_dep_time % 100, dep_delay_val = lambda x : x.dep_time_min - x.sched_dep_time_min  )
```

### Brainstorm at least 5 different ways to assess the typical delay characteristics of a group of flights. Consider the following scenarios:

#### A flight is 15 minutes early 50% of the time, and 15 minutes late 50% of the time.

```python
not_canceled.groupby('flight').agg( 
    arr_delay_early_50 = ('arr_delay', np.median)
).query('arr_delay_early_50 == 15').head(5)
```

```python
not_canceled.groupby('flight').agg( 
    arr_delay_early_50 = ('arr_delay', np.median)
).query('arr_delay_early_50 == -15').head(5)
```

#### A flight is always 10 minutes late.

```python
not_canceled.groupby('flight').agg(
    avg = ('arr_delay', np.mean),
    std = ('arr_delay', np.std)
).query('avg == 10')
```

#### A flight is 30 minutes early 50% of the time, and 30 minutes late 50% of the time.

```python
not_canceled.groupby('flight').agg( 
    arr_delay_early_50 = ('arr_delay', np.median)
).query('arr_delay_early_50 == 30').head(5)
```

```python
not_canceled.groupby('flight').agg( 
    arr_delay_early_50 = ('arr_delay', np.median)
).query('arr_delay_early_50 == -30').head(5)
```

#### 99% of the time a flight is on time. 1% of the time it’s 2 hours late.

```python
not_canceled.groupby('flight').agg(
    arr_delay_99 = ('arr_delay', lambda x : np.quantile(x, 0.99)),
    median = ('arr_delay', np.max)
).query('arr_delay_99 == 0')
```

```python
not_canceled.groupby('flight').agg(
    arr_delay_99 = ('arr_delay', lambda x : np.quantile(x, 0.01))
).query('arr_delay_99 == 0')
```
