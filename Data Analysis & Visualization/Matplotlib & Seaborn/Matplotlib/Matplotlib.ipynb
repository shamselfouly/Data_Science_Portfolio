### Functional Method to Create Plots

```python
x = np.linspace(0,5,11)
y = x ** 2

#Functional Method of creating plots
plt.plot(x,y, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('String title here')
```

### Subplots Using Functional Method

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### OOP Method - Single Plot

```python
fig = plt.figure() #Created a figure called fig. An imaginary blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Calling add_axes method off of figure object (Left, bottom, width & height)

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4, 0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figure1')

axes2.plot(y,x)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Figure2')
```

### OOP Method - Subplots 

```python
#Automatically creates axes for us unlike manually creating it above from nrows & ncols
#Combines both fig & axes lines of code above
fig,axes = plt.subplots(nrows = 1, ncols = 2) 

#We get an array of axis
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Figure1')

axes[1].plot(y,x)
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Figure2')

plt.tight_layout() #fixes overlaps
```

## Graph Apprearance

#### Defining Figure Size & & DPI - Single Graph

```python
fig = plt.figure(figsize = (8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

#### Defining Figure Size & & DPI - Subplots

```python
#figsize with Subplots
fig,axes = plt.subplots(2,1,figsize = (8,4), dpi = 80)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

#### Adding Legends 

```python
#Adding legends to graph
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")
#Chooses the best locatin for my legend
ax.legend(loc=0)
```

#### Plot styling - Adding Colors, transprancy, line & marker characristics & axis limits

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
#RGB Hex codes
#Alpha controls transprancy 
#linewidth = lw / linestyle = ls / markersize = ms
#markers mark where data points are 
ax.plot(x,y, color = '#FF8C00', lw = 3, alpha = 1, ls = '-', marker = 'o', ms = 10,
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Figure1')
ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

#Set x & y limits
ax.set_xlim([0,6])
ax.set_ylim([0,30])
```

#### Saving and exporting figure with desired DPI

```python
#Saving plot
fig.savefig('my_picture.png', dpi = 200)
```
