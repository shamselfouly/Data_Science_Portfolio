```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Grids (Used to get more customization power)

## PairGrid

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
