```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Part 1 (Distribution Plots)
1. Histogram Plot (sns.histplot())
2. kdeplot (sns.kdeplot())
3. Jointplot (sns.jointplot())
4. pairplot (sns.pairplot())
5. rugplot (sns.rugplot())

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Distplot
    # Distrubtion of a univariate set of observations (i.e. one variable)
    # Histogram & KDE (Kernal Density estimation)
```

```python
sns.distplot(tips['total_bill'], kde = True, bins = 20)
```

```python
#Joint plots
    # Allows you to match up two histplots for bivariant data
    # Shows the two distribution plots and and a middle plot which I can choose with the kind keyword
    # hex, reg, kde, resid & scatter(Default)
    # kde: the density where the points match up the most
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind = 'kde')
```

```python
#Pairplots
    # Nice way to quickly visualize data
    # joinplot for every single numerical column in the dataset and hue introduces categorical data
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```
