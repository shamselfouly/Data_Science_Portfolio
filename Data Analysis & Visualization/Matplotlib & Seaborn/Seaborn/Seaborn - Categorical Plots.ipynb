```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Part 2 (Categorical Plots) x is always categorical

1. Barplot (includes the pandas version as well)
2. Countplot (includes the pandas version as well)
3. boxplot
4. Volinplot
5. Stripplot
6. Swarmplot
7. Swarm + Violin Plot
8. catplot or factorplot (General plot to call most of them)

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#x is categorical and y is a numerical column
sns.barplot(x='sex', y='total_bill' ,data=tips)
```

```python
tips.groupby('sex')['total_bill'].mean()
```

```python
#Similar to a barplot instead that the estimator is set to count number of occurances. 
#Similar to pandas value_count method (See below)
sns.countplot(x = 'sex', data = tips)
```

```python
tips['sex'].value_counts()
```

```python
# Distribution of categorical data as oppose to distrubtion of numerical data
sns.boxplot(x='day',y='total_bill',data=tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data=tips, hue = 'sex', split = True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips,kind='strip')
```
